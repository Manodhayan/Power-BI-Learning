```python
import urllib, time
import urllib.request as urllib2
from datetime import datetime

# REST API endpoint, given to you when you create an API streaming dataset
# Will be of the format: https://api.powerbi.com/beta/<tenant id>/datasets/< dataset id>/rows?key=<key id>
REST_API_URL = "https://api.powerbi.com/beta/3f80d8b1-b9fa-4081-8625-ff2174620d6f/datasets/2b3e4ff0-277e-40cb-bd2a-14bc490acdfb/rows?key=IUXHQN%2FP9EEN%2BCSKLOm5s%2FUJ9HNxgoSrYhpS5bm6WQMMNGRqSzrJ384TbRGyGDKD5JrRE%2BQWqe7xWCb%2F%2FGO3%2Bw%3D%3D"

import pandas as pd
from datetime import datetime
from datetime import timedelta
import requests
import time
import random

##class for data_generation


def data_generation():
    time = datetime.now()
    temp=random.uniform(24, 28)
    hum=random.uniform(58, 65)
    room='ICU'

    return [time, hum, temp, room]


if __name__ == '__main__':
    while True:
        data_raw = []
        for i in range(1):
            row = data_generation()
            data_raw.append(row)
            print("Raw data - ", data_raw)

        # set the header record
        HEADER = ["Timestamp", "Humidity", "Temperature", "Room"]

        data_df = pd.DataFrame(data_raw, columns=HEADER)
        data_json = bytes(data_df.to_json(orient='records'), encoding='utf-8')
        print("JSON dataset", data_json)

        # Post the data on the Power BI API
        req = requests.post(REST_API_URL, data_json)

        print("Data posted in Power BI API")
        time.sleep(2)
```
